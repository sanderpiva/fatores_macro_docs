```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

data_fim = datetime(2025, 10, 2)
data_inicio = data_fim - timedelta(days=5 * 365 + 1)

tickers = ['PETR4.SA', 'VALE3.SA', 'ITUB4.SA']

dados_acoes = yf.download(
    tickers,
    start=data_inicio,
    end=data_fim,
    interval='1d'
)

try:
    # 4.1. Tenta selecionar o preço de fechamento ajustado (o ideal para análise)
    precos_finais = dados_acoes['Adj Close']
    print("Preços baixados e ajustados (Adj Close) com sucesso.")
except KeyError:
    # 4.2. Se der erro (KeyError), usa o preço de Fechamento Nominal como alternativa
    print("Coluna 'Adj Close' não encontrada. Usando 'Close' como alternativa.")
    precos_finais = dados_acoes['Close']

precos_finais.columns = ['ITUB4', 'PETR4', 'VALE3']

print("\n--- Visualização Limpa do Preço de Fechamento ---")
print(precos_finais.head())

precos_finais.to_csv('precos_acoes_5anos.csv')
```
